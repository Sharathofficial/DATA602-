#### Question 2: Wide data isn't usually helpful.
Convert this to a long format with the following columns:  
1. country  
2. year  
3. diagnosis method: mapping should be: 

  {'newsp':'Pulmonary Smear (+)'
  ,'newsn':'Pulmonary Smear (-)'
  ,'newep':'Extrapulmonary'
  ,'newrel':'Relapse'}   
  

4. gender: male or female.   
5. age: lower_age - higher_age, e.g., 0-14  
6. number of cases  

```python
# Code
df_new = df.melt(id_vars=['country','year'],var_name=['new_sp'],value_name='TB_cases')
df_new
```

```python
import re

df_new['diagnosis method'] = df_new['new_sp'].apply(lambda x:'Pulmonary Smear (+)' if re.search(r'new_sp',x) else ('Pulmonary Smear (-)' if re.search(r'new_sn',x) else ('Extrapulmonary' if re.search(r'newep',x) else 'Relapse')))
df_new
```

```python
df_new['Sex'] = df_new['new_sp'].apply(lambda x:'Male' if re.search(r'_m',x) else 'Women' )
df_new
```

```python
df_new['Age'] = df_new.new_sp.str.extract('(\d+)')
df_new
```

```python
ageMap = {'014' :'0-14',
            '1524':'15-24',
           '2534':'25-34', 
           '4554':'45-54', 
           '5564':'55-64', 
           '65':'65'
          }
df_new['AGE_CAT'] = df_new['Age'].map(ageMap)
```

```python
df_new
```

```python
df_new = df_new.drop(columns=['new_sp', 'Age'])
df_new.info()
```

#### Question 3: Create a graph that shows when countries started to report TB cases.

```python
# Code
countrylist = list(df_new.country.unique())
rangelen = range(0,len(countrylist)) 
yearlist = list(df_new.year.unique())
```

```python
country_year  = df_new.query('TB_cases> 0.0').groupby('country')['year'].min()
```

```python
fig= plt.figure(figsize=(25,30))
country_year.plot(kind='line')
plt.xticks(ticks = rangelen ,labels = countrylist, rotation = 'vertical')
plt.yticks(yearlist)
plt.xlabel('Countries')
plt.ylabel('Year')
plt.title('Years of Countries started to report TB cases')
plt.show()
```
